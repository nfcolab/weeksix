This notebook covers how to use the decision making model to investigate the effect of colors. Note that in addition to `ui.py`, you need `decision.py`.

```python
import ui
import decision
```

The decision making model is implemented as an instance of the `ui` class, named `decision_ui` (remember to add the prefix `decision.` to access it, because it comes from `decision.py`. In addition, there is a helper function for setting the UI for a given task, two tasks ready-made, and a function for running a decision making simulation on a given UI.

```python
# This is the UI without any task:
ui.visualise_UI(decision.decision_ui)
```

Tasks are defined as dictionaries. Here is the specification of the first task:

`element_values1 = {
    "e1": -0.3,
    "e2": 0.3,
    "e3": 0.2,
    "e4": 0.3,
    "e5": -0.3,
    "e6": 0.3,
    "e7": -0.2,
    "e8": 0.5,
    "e9": -0.4
    }
`

For each element in the UI, it provides a number. The model makes its decision on the basis of encoding these numbers. We can init the decision UI with these element values like this:

```python
my_values = {
    "e1": -0.3,
    "e2": 0.3,
    "e3": 0.2,
    "e4": 0.3,
    "e5": -0.3,
    "e6": 0.3,
    "e7": -0.2,
    "e8": 0.5,
    "e9": -0.4
    }
```

```python
my_values['e5']
```

```python
decision.reset_decision_ui(decision.decision_ui, my_values)
ui.visualise_UI(decision.decision_ui)
```

As you can see, the correct values have now been set. You can try to change `element_values1` to `element_values2` in order to check that it works as well.

Here is how you add colour to help the decision maker:

```python
decision.reset_decision_ui(decision.decision_ui, my_values, color = True)
ui.visualise_UI(decision.decision_ui)
```

Be sure to know what values and features the UI has been initialised before running tasks with it. Here is the model running a decision making task on the UI just visualised above.

Make sure that `decision.decision_ui` has been set to the same task (`element_values`) as the parameter given to the `make_decision` function. The next cell assumes that you have run `decision.reset_decision_ui(decision.decision_ui, decision.element_values1, color = True)` (with the `color` parameter either way, of course).

```python
decision.make_decision(decision.decision_ui, my_values)[1]
```
